```python
from datasets import load_dataset, Dataset
dataset = load_dataset("IlyaGusev/ru_turbo_alpaca", trust_remote_code=True, split="train")
print(dataset[0])
print(dataset[2000])
```

```python
dataset = dataset.filter(lambda x: x.get("label") != "bad_output")
dataset[0]
```

```python
def format_instruction(example):
    return {
        "text": f"""<s>[INST] <<SYS>>Ты — русскоязычный ассистент. Отвечай точно и профессионально.<</SYS>>
{example['instruction']}
{example['input']} [/INST] {example['output']}</s>"""
    }

dataset = dataset.map(format_instruction, remove_columns=dataset.column_names)
```
